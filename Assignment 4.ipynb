```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
#Reading CSV file
data=pd.read_csv('/content/dat2.csv',encoding="ISO-8859-1")
print(data.to_string())
```

```python
#Finding duplicate entries the Duplicate entries
print((data.duplicated()).to_string())
```

```python
#Droping the Duplicate Data
data.drop_duplicates(inplace=True)
print(data.to_string())
```

```python
#removing the rows having NaN values
data.dropna(inplace=True)
print(data.to_string())
```

```python
#Setting same format for Date column
data['Date']=pd.to_datetime(data['Date'])
print(data.to_string())
```
